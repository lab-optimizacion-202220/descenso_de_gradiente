# Primer punto

## 1) Algoritmo descenso de gradiente.

```python
# función

def gam(x):
    ans = ((x[0])**2)  +(2*(x[1])**2) + (0.5*x[0]) + 5 
    return ans 

def des_grad (xo,n,alpha,umb):
    x = xo
    vx = []
    vy = []
    i = 0
    r = 10
    while (i<= n and (r >= umb and r <= 200)):
        print('------------------------')
        print('iteración ', str(i+1))
        # Calcular gradiente
        gradiente = nd.Gradient(gam)(x)
        valor = gam(x)
        # Actualizar "x" usando gradiente descendente
        xaux = x
        x = x - alpha*gradiente
        r = ((xaux[0]-x[0])**2+(xaux[1]-x[1])**2)**0.5
        # Almacenar iteración y valor correspondiente
        vx.append(x[0])
        vy.append(x[1])
        # Imprimir resultados
        print('x = ', str(x[0]), ', y = ', str(x[1]), "z = ",str(valor), "r = ", str(r))
        i += 1
    return vx,vy
```

```python
def gam(x):
    ans = ((3*(1-x[0])**2)*e**(-(x[0]**2)-(x[1]+1)**2)) -(10*(0.2*x[0]-x[0]**3-x[1]**5)*e**(-x[0]**2 -x[1]**2)) - (0.333*e**(-(x[0]+1)**2 -x[1]**2)) 
    return ans 

def des_grad (xo,n,alpha,umb):
    x = xo
    vx = [xo[0]]
    vy = [xo[1]]
    i = 0
    r = 10
    while (i<= n and (r >= umb and r <= 500)):
        print('------------------------')
        print('iteración ', str(i+1))
        # Calcular gradiente
        h = Symbol('h')
        d1 = [x[0]+h, x[1]]
        d2 = [x[0], x[1]+h]
        gradiente = np.array([0,0])
        gradiente[0] = limit((gam(d1) - gam(x))/h, h, 0.00000000000001)
        gradiente[1] = limit((gam(d2) - gam(x))/h, h, 0.00000000000001)
        valor = gam(x)
        # Actualizar "x" usando gradiente descendente
        xaux = x
        x = x - alpha*gradiente
        r = ((xaux[0]-x[0])**2+(xaux[1]-x[1])**2)**0.5
        # Almacenar iteración y valor correspondiente
        vx.append(x[0])
        vy.append(x[1])
        # Imprimir resultados
        print('x = ', str(x[0]), ', y = ', str(x[1]), "z = ",str(valor), "r = ", str(r))
        i += 1
    return vx,vy

xo2= [-0.5,-1.2]
alpha = 0.01
n= 200
r = 0.00005
vx,vy=des_grad (xo2,n,alpha,r)
x1 = np.linspace(-5,5,500)
x2 = np.linspace(-5,5,500)
X1,Y1 = np.meshgrid(x1,x2)

# A1
plt.figure(4)
plt.title("Contorno para f(x) con xo= [-0.5,-1.2]")

Z =((3*(1-X1)**2)*e**(-(X1**2)-(Y1+1)**2)) -(10*(0.2*X1-X1**3-Y1**5)*e**(-X1**2 -Y1**2)) - (0.333*e**(-(X1+1)**2 -Y1**2)) 
ax=plt.contour(X1,Y1,Z,50)
plt.plot(vx,vy,"-o",color="red")
plt.colorbar(ax,shrink=0.5,aspect=4)
```

```python
def gam(x):
    ans = 3*x[0]**2 + 2*x[1]**2 + 4*x[0] + 5*x[1]  
    return ans 

def des_grad (xo,n,alpha,umb):
    x = xo
    vx = [xo[0]]
    vy = [xo[1]]
    i = 0
    r = 10
    while (i<= n and (r >= umb and r <= 500)):
        print('------------------------')
        print('iteración ', str(i+1))
        # Calcular gradiente
        gradiente = nd.Gradient(gam)(x)
        valor = gam(x)
        # Actualizar "x" usando gradiente descendente
        xaux = x
        x = x - alpha*gradiente
        r = ((xaux[0]-x[0])**2+(xaux[1]-x[1])**2)**0.5
        # Almacenar iteración y valor correspondiente
        vx.append(x[0])
        vy.append(x[1])
        # Imprimir resultados
        print('x = ', str(x[0]), ', y = ', str(x[1]), "z = ",str(valor), "r = ", str(r))
        i += 1
    return vx,vy

```
